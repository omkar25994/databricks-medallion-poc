```python
spark.sql("""
CREATE OR REPLACE TABLE gold.suzuki_summary AS
SELECT 
    year(Date) as year,
    month(Date) as month,
    AVG(Close) as avg_close,
    SUM(Volume) as total_volume,
    MAX(High) as max_price,
    MIN(Low) as min_price
FROM new.default.suzuki_stock_silver
GROUP BY year(Date), month(Date)
""")
```
